## Convert Pixel format to YOLO 0-1 normalized coordinates.

```python
import os
import cv2
import re
from pathlib import Path
from PIL import Image

images_path = Path('/Users/eashan22/Dropbox (LU Student)/Macbook/Desktop/Media Forensics/mini-project-2/Task 1/data/test')
anno_path = Path('/Users/eashan22/Dropbox (LU Student)/Macbook/Desktop/Media Forensics/mini-project-2/Task 1/data/test/test_bbox.txt')

imageList, labelList, anno_list = [], [], []
imageList = os.listdir(images_path)
imageList.remove('test_bbox.txt')
sorted_imageList = sorted(imageList)

labelList = open(anno_path, "r")
labels = labelList.readlines()

def fix_renormalize(xmin, xmax, ymin, ymax):
    if xmin<0.0:
        xmin = 0.0
    elif xmin>1.0:
        xmin = 1.0

    if ymin<0.0:
        ymin = 0.0
    elif ymin>1.0:
        ymin = 1.0

    if xmax<0.0:
        xmax = 0.0
    elif xmax>1.0:
        xmax = 1.0
    
    if ymax<0.0:
        ymax = 0.0
    elif ymax>1.0:
        ymax = 1.0
    return xmin, ymin, xmax, ymax

def convert(size, box):
    """
    https://stackoverflow.com/questions/64096953/how-to-convert-yolo-format-bounding-box-coordinates-into-opencv-format
    https://github.com/ivder/LabelMeYoloConverter/blob/master/convert.py
    """
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x, y, w, h)


for image, label in zip(sorted_imageList,labels):
    image_label_idx = [re.findall(r'(\w+?)(\d+)', image)[0]][0][1]
    filename = "/Users/eashan22/Dropbox (LU Student)/Macbook/Desktop/Media Forensics/mini-project-2/Task 1/data/test/"+image[:-9]+image_label_idx+".png"
    lab = label.split(" ")
    f = Path(filename)
    
    print(filename)
    im = Image.open(filename) # cv2.imread(filename)
    w = int(im.size[0])
    h = int(im.size[1])

    x1, y1, x2, y2, class_ = lab[0], lab[1], lab[2], lab[3], lab[4]
    if class_.split("\n")[0] == "tamper":
        class_label = 1
    elif class_.split("\n")[0] == "authentic":
        class_label = 0

    x1, y1, x2, y2 = int(float(x1)), int(float(y1)), int(float(x2)), int(float(y2))
    xmin = min(x1, x2)
    xmax = max(x1, x2)
    ymin = min(y1, y2)
    ymax = max(y1, y2)

    # print("Before: ", x1, y1, x2, y2)
    b = (xmin, xmax, ymin, ymax)
    xmin, ymin, xmax, ymax = convert((w,h), b)
    xmin, ymin, xmax, ymax = fix_renormalize(xmin, ymin, xmax, ymax)

    # print("After :", xmin, ymin, xmax, ymax)
    
    anno = (class_label, xmin, ymin, xmax, ymax)
    anno_ = "{} {:.05f} {:.05f} {:.05f} {:.05f}".format(*anno)

    yolofile = open("/Users/eashan22/Dropbox (LU Student)/Macbook/Desktop/Media Forensics/mini-project-2/Task 1/data/anno/test/"+f.parts[-1][:-4]+".txt", "w")
    yolofile.write( '%s\n' % str(anno_))#[1:-1])
    yolofile.close()

print("Done")
```
